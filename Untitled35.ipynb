```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )


    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
```

The target of this function is to create a correlation heatmap using seaborn First of all the .corr() method will actually get rid of the columns that are not suited for correlation for you.If you wanted less categories in your heat map,you could select only those ones ,in other words the purpose of .corr() is to compute the pairwise correlation of columns excluding NaN values.Then we created the size for a plot to do that we use the plt.seaborn and we put in a figure size of twelve by ten.Furthermore, sns.diverging_palette Make a diverging palette between two HUSL colors that will be used in the graph.Then we use seaborn heatmap :"sns.heatmap" in our dataframe with our correlation we want to go ahead and add the annotations from our correlation matrix and for our color map.
